```python
import numpy as np
import random as rnd
```

```python
target= "hello world"
population = 1000
treshold_mutation = 0.10
treshold_worst = 0.80
treshold_parent = 0.10
```

```python
letters = "qwertyuiopasdfghjklñzxcvbnm" #letras del abecedario
```

```python
#inicializar población
words=[]

for j in range (population):
    word = ""
    for i in range (len(target)):
        word += rnd.choice(letters)
    words.append([0, word])
```

```python
for entry in words:
    print(entry)
```

```python
def fitness(words,target):
    for entry in words:
        word = entry[1]
        score = 0
        for i in range(len(word)):#Por cada caracter correcto aumentamos la puntuación de la palabra
            if word[i] == target[i]:
                score += 1
        entry[0] = score
```

```python
fitness(words, target)
```

```python
for entry in words:
    print(entry)
```

```python
words = sorted(words,reverse = True)#Python eleige el primer elemento de la lista para ordenar listas

for entry in words:
    print(entry)
```

```python
index = int (population*treshold_worst)#Porcentaje a eliminar

for i in range (index, population):#Elimnar los peores elementos y resetar la lista
    words[i] = [0,'']
```

```python
def crossover (parent1, parent2):
    child = ""
    for i in range(len(parent1)):
        choice = rnd.random()
        if choice > 0.5:
            child += parent1[i]
        else:
            child += parent2[i]
        return child
```

```python
crossover("1234567890","abcdefghijk")
```

```python
index = int (population*treshold_parent)
parents = words[:index]
```

```python
index = int (population*treshold_worst)
for i in range(index, population):
    parent1 = rnd.choice(parents)
    parent2 = rnd.choice(parents)
    child = crossover(parent1[1],parent2[1])
    words[i][1] = child
```

```python
def mutation(word):
    choice = rnd.random()
    if choice < treshold_mutation:
        index = rnd.randint(0,len(word)-1)
        word = word[:index] + rnd.choice(letters) + word[index+1:]
    return word
```
