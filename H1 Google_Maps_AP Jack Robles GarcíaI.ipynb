```python
#Mas sencillo que rezar a google maps que te devuela lo que quieres
def BuscarLugaresPorNombre(tipoLocal,radioDeBusqueda, longitud,latitud, llave):
    radioDeBusqueda=radioDeBusqueda*1000
    
    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={}&inputtype=textquery&fields=name,rating,place_id,types,user_ratings_total,geometry&locationbias=circle{}@{},{}&key={}'.format(tipoLocal,radioDeBusqueda, longitud,latitud, llave)

    res = rq.get(url)
    jason = json.loads(res.content)
    return jason
```

```python
StbKJs = BuscarLugaresPorNombre('Starbucks',1,20.994484, -89.614743, llave,None)
Stb = pd.read_json(json.dumps(StbKJs['candidates']))
Stb['lat'] = StbKJs['candidates'][0]['geometry']['location']['lat']
Stb['lon'] = StbKJs['candidates'][0]['geometry']['location']['lng']
Stb
```

```python
#
centroide = (20.974266, -89.621795)

def findSimilar(nombre,key,rango,centro):
    '''
    rango en KM
    centro: tupla de cordenadas del centro de busqueda(lon,lat)
    '''
    guardaJsons = BuscarLugaresPorNombre(nombre,20,centro[0], centro[1], key)
    lat = guardaJsons['candidates'][0]['geometry']['location']['lat']
    lon = guardaJsons['candidates'][0]['geometry']['location']['lng']
    types = guardaJsons['candidates'][0]['types'][0]
    
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&type={}&radius={}&rankby=prominence&key={}'.format(lat,lon,types,(rango*1000),key)
    print(url)
    
    res = rq.get(url)
    guardaJsons = json.loads(res.content)
    return guardaJsons
```

```python
preDF4 = findSimilar('Starbucks',llave,20,centroide)
preDF4
```
